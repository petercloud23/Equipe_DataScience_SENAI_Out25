```python
'''Lembretes importantes!!!
Listas → boas para armazenar coleções simples.

Dicionários → ótimos para estruturar dados com pares chave → valor.

pandas DataFrame → ideal quando você precisa de uma tabela profissional para análise, relatórios ou integração com ferramentas de ciência de dados.'''

'''Vamos evoluir o exemplo anterior, da média dos alunos para que além de mostrar as notas e médias na tela, o programa também armazene as médias
em uma nova lista (medias).
Assim, você pode reutilizar esses valores em relatórios, comparações ou até mesmo em análises posteriores.'''

# Lista de alunos
alunos = ["Ana", "João", "Maria"]


# Matriz com as notas (cada linha representa um aluno e cada coluna uma prova)
notas = [
    [7.5, 8.0, 9.0],  # notas da Ana
    [6.0, 5.5, 7.0],  # notas do João
    [9.0, 8.5, 10.0]  # notas da Maria
]

# Lista vazia para armazenar as médias
medias = []

# Percorrer alunos e calcular suas médias
for i in range(len(alunos)):
    media = sum(notas[i]) / len(notas[i])  # calcula média
    medias.append(round(media,2))  # armazena a média arredondada ('round') na lista, essa função evolui o exemplo anterior criando "memória"

    # Exibe as notas e a média
    print(f"Notas de {alunos[i]}: {notas[i]}")
    print(f"Média de {alunos[i]}: {round(media,2)}")
    print("------")
```

```python
# Podemos consultar e exibir a lista de médias final posteriormente, sem executar todo o código anterior, pois agora estão salvas
print("Lista de médias: ", medias)
```

```python
'''Um dicionário (dict) é uma estrutura de dados que armazena pares chave → valor.

Exemplo simples:

aluno = {"nome": "Ana", "notas": [7.5, 8.0, 9.0], "media": 8.17}

Acima, "nome", "notas" e "media" são chaves, e os valores associados podem ser números, listas, textos, etc.

Isso facilita muito quando queremos acessar informações pelo nome da chave em vez de depender de índices de lista."'''

# Vamos implementar o exemplo anterior da media com Dicionário

# Lista de alunos
alunos = ["Ana", "João", "Maria"]

# Matriz com notas
notas = [[7.5, 8.0, 9.0],[6.0, 5.5, 7.0],[9.0, 8.5, 10.0]]

# Criar lista de dicionários
dados_alunos = []

for i in range(len(alunos)):
    media = sum(notas[i]) / len(notas[i])
    aluno_dict = {
        "nome": alunos[i],
        "notas": notas[i],
        "media": media
    }
    dados_alunos.append(aluno_dict)

```

```python
print(dados_alunos)
```

```python
# Exibindo
for aluno in dados_alunos:
    print(f"{aluno['nome']} - Notas: {aluno['notas']} - Média: {aluno['media']:.2f}")
```

```python
# Exibindo a média de um aluno específico diretamente, desde que pré-cadastrado
maria_data = None
nome = input("Digite o nome do aluno(a)")
nome = nome.lower().strip()
for aluno in dados_alunos:
    if aluno["nome"].lower() == nome:
        maria_data = aluno
        break

if maria_data:
    print(f"Média de {maria_data['nome']}: {maria_data['media']:.2f}, baseado nas notas: {maria_data['notas']}")
else:
    print("Aluno não encontrado.")
```

```python
'''DataFrame no pandas

Um DataFrame é uma estrutura de dados fornecida pela biblioteca pandas, muito parecida com uma tabela do Excel.

Ele organiza dados em linhas e colunas, permitindo operações poderosas de análise, filtragem, ordenação, cálculos estatísticos, etc.

É amplamente usado em ciência de dados, estatística e aprendizado de máquina.'''

#Exemplo com pandas DataFrame

import pandas as pd

# Criando DataFrame a partir da lista de dicionários
df = pd.DataFrame(dados_alunos)

# Exibindo o DataFrame
print(df)
```

```python
# Exemplo: selecionar só nomes e médias
print("\nRelatório resumido:")
print(df[["nome", "media"]])
```

# Exemplo prático: Capturar temperaturas de uma fonte externa e calcular a média em determinado horário
Primeiro, vamos instalar a biblioteca pandas, caso você ainda não a tenha instalado.

```python
import pandas as pd
```

Agora, vamos criar o código para ler o arquivo CSV, obter as informações necessárias do usuário e calcular a média da temperatura para a hora especificada.

```python
# Obter o nome do arquivo CSV do usuário
file_name = input("Digite o nome do arquivo CSV (incluindo a extensão, por exemplo, temperaturas.csv): ")

# Obter o número de dias no mês do usuário
num_dias = int(input("Digite o número de dias no mês (por exemplo, 30 ou 31): "))

# Obter a hora para a qual calcular a média (lembre-se que as colunas são as horas)
hora_para_media = input("Digite o nome da coluna (hora) para a qual você quer calcular a temperatura média: ")

try:
    # Carregar o arquivo CSV em um DataFrame pandas
    df_temperaturas = pd.read_csv(file_name)

    # Verificar se a coluna da hora especificada existe no DataFrame
    if hora_para_media not in df_temperaturas.columns:
        print(f"Erro: A coluna '{hora_para_media}' não foi encontrada no arquivo CSV.")
    else:
        # Selecionar os dados da coluna da hora especificada para o número de dias informado
        # Assumimos que as primeiras 'num_dias' linhas correspondem aos dias do mês
        temperaturas_hora_especifica = df_temperaturas[hora_para_media].head(num_dias)

        # Calcular a média das temperaturas para a hora especificada
        media_temperatura = temperaturas_hora_especifica.mean()

        # Imprimir o resultado
        print(f"\nA temperatura média para a hora '{hora_para_media}' durante {num_dias} dias é: {media_temperatura:.2f}")

except FileNotFoundError:
    print(f"Erro: O arquivo '{file_name}' não foi encontrado. Verifique o nome e o caminho do arquivo.")
except ValueError:
    print("Erro: Certifique-se de digitar um número válido para o número de dias.")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```
