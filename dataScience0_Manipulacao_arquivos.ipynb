````python
# ==============================================================================
# Exemplos de Manipulação de Arquivos em Python (Nível Iniciante)
# ==============================================================================

# Exemplo 1: Escrever em um arquivo de texto e depois ler seu conteúdo

# Define o nome do arquivo
nome_arquivo = "teste de aula1.txt" #Este arquivo existe, por enquanto, somente dentro do ambiente de execução

# Modo 'w' (write): Abre o arquivo para escrita. Se o arquivo existir,
#seu conteúdo será apagado. Se não existir, será criado.
# O 'with open(...)' garante que o arquivo será fechado automaticamente,
# mesmo que ocorra um erro.
with open(nome_arquivo, "w", encoding="utf-8") as arquivo:
    # Escreve linhas no arquivo. O caractere '\n' representa uma quebra de linha.
    arquivo.write("Nome: Fulano de tal\n")
    arquivo.write("Idade: 21 anos\n")
    arquivo.write("Profissão: Estudante\n")

print(f"Conteúdo escrito em '{nome_arquivo}'")
````

````python
# Versão do código acima que solicita o nome do arquivo e
# conteúdo ao usuário, via função input
nome_arquivo2 = input("Digite o nome do arquivo") +".txt"

with open(nome_arquivo2, "w") as arquivo2:
    arquivo2.write(input("Digite o conteúdo do arquivo: "))
print(f"Conteúdo escrito em '{nome_arquivo2}'")
````

````python
# Modo 'r' (read): Abre o arquivo para leitura.
with open(nome_arquivo2, "r", encoding="utf-8") as arquivo:
    # Lê todo o conteúdo do arquivo.
    conteudo = arquivo.read()
    print(f"\nConteúdo lido de '{nome_arquivo2}':")
    print()
    print(conteudo)
````

````python
# Exemplo 2: Adicionar conteúdo a um arquivo existente

# Modo 'a' (append): Abre o arquivo para escrita.
#Se o arquivo existir, o novo conteúdo será adicionado ao final.
#Se não existir, será criado.
with open(nome_arquivo2, "a", encoding="utf-8") as arquivo:
    # Adiciona mais uma linha ao final do arquivo.
    arquivo.write("Nome: José Maria\n")
    arquivo.write("Idade: 23 anos\n")
    arquivo.write("Profissão: Diretor\n")

print(f"Conteúdo adicionado a '{nome_arquivo2}'")
````

````python
# Lê novamente o conteúdo para ver a linha adicionada
with open(nome_arquivo2, "r", encoding="utf-8") as arquivo:
    conteudo_atualizado = arquivo.read()
    print(f"\nConteúdo atualizado de '{nome_arquivo2}':")
    print(conteudo_atualizado)
````

````python
# Para baixar para a máquina ou servidor local o arquivo criado dentro do ambiente de execução,
# permitindo ao usuário selecionar o arquivo.

from google.colab import files
import os

# Obtém a lista de arquivos no diretório atual (normalmente /content/ no Colab)
# Filtra para incluir apenas arquivos (excluindo diretórios)
arquivos_disponiveis = [f for f in os.listdir('.') if os.path.isfile(f)]

if not arquivos_disponiveis:
    print("Não há arquivos disponíveis para download no diretório atual.")
else:
    print("Arquivos disponíveis para download:")
    # Mostra a lista de arquivos numerada para o usuário
    for i, arquivo in enumerate(arquivos_disponiveis):
        print(f"{i + 1}: {arquivo}")

    try:
        # Solicita ao usuário que escolha um arquivo
        escolha = input(f"Digite o número do arquivo que deseja baixar (1 a {len(arquivos_disponiveis)}): ")
        indice_escolhido = int(escolha) - 1 # Converte a escolha para índice baseado em 0

        # Verifica se o índice escolhido é válido
        if 0 <= indice_escolhido < len(arquivos_disponiveis):
            nome_do_arquivo_para_baixar = arquivos_disponiveis[indice_escolhido]

            try:
                files.download(nome_do_arquivo_para_baixar)
                print(f"Arquivo '{nome_do_arquivo_para_baixar}' solicitado para download.")
            except FileNotFoundError:
                print(f"Erro: Arquivo '{nome_do_arquivo_para_baixar}' não encontrado no ambiente Colab (apesar de estar na lista).")
            except Exception as e:
                print(f"Ocorreu um erro durante o download: {e}")
        else:
            print("Escolha inválida. Por favor, digite um número da lista.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
    except Exception as e:
        print(f"Ocorreu um erro inesperado durante o processo de seleção: {e}")
````

O trecho de código: arquivos_disponiveis = [f for f in os.listdir('.') if os.path.isfile(f)] é um exemplo de uma list comprehension (compreensão de lista) em Python.

Ele faz o seguinte:

os.listdir('.'): Esta parte lista todos os itens (arquivos e diretórios) presentes no diretório atual (representado por '.').
for f in os.listdir('.'): Isso itera sobre cada item retornado por os.listdir('.'). Para cada item, ele o atribui temporariamente à variável f.
if os.path.isfile(f): Esta é uma condição. Para cada item f, a função os.path.isfile(f) verifica se f é um arquivo (e não um diretório ou outro tipo de item do sistema de arquivos). A condição é verdadeira se for um arquivo.
f: Se a condição if os.path.isfile(f) for verdadeira, o item f (que é o nome do arquivo) é incluído na nova lista que está sendo criada.
[...]: Os colchetes indicam que o resultado de toda a operação é uma nova lista.
Portanto, a linha inteira constrói uma nova lista chamada arquivos_disponiveis que contém apenas os nomes dos arquivos (excluindo diretórios) presentes no diretório atual. Isso é útil para garantir que a lista mostrada ao usuário e usada para download contenha apenas itens que podem ser baixados como arquivos.

````python
# Exemplo de como excluir um arquivo usando os.remove()

import os

nome_arquivo_para_excluir = input("Digite o nome do arquivo a excluir: ")+".txt" # escolha nome do arquivo que deseja excluir

try:
    # Exclui o arquivo
    os.remove(nome_arquivo_para_excluir)
    print(f"Arquivo '{nome_arquivo_para_excluir}' excluído com sucesso.")
except FileNotFoundError:
    print(f"Erro: Arquivo '{nome_arquivo_para_excluir}' não encontrado para exclusão.")
except Exception as e:
    print(f"Ocorreu um erro ao tentar excluir o arquivo: {e}")
````

````python
# Permite ao usuário escolher o modo de abertura do arquivo ('w', 'r' ou 'a')
# Encapsulado em uma função

import os
from google.colab import files # Import files para download

def manipular_arquivo():

    nome_arquivo =input("Digite o nome do arquivo") + ".txt" # Usando o nome de arquivo já definido anteriormente

    # Solicita ao usuário o modo desejado ou ação
    modo_escolhido = input(f"Escolha o modo para abrir '{nome_arquivo}' ('w' para escrever, 'r' para ler, 'a' para adicionar, 'd' para download, 'x' para excluir): ").lower()

    # --- Camada Simples de Cibersegurança ---
    # Validar a entrada do usuário para garantir que é um modo permitido.
    modos_permitidos = ['w', 'r', 'a', 'd', 'x']

    if modo_escolhido not in modos_permitidos:
        print(f"Modo '{modo_escolhido}' inválido. Escolha entre {modos_permitidos}.")
    else:
        try:
            if modo_escolhido in ['w', 'r', 'a']:
                # Abre o arquivo nos modos de manipulação de conteúdo
                with open(nome_arquivo, modo_escolhido, encoding="utf-8") as arquivo:
                    print(f"\nArquivo '{nome_arquivo}' aberto no modo '{modo_escolhido}'.")

                    if modo_escolhido == 'r':
                        # Modo leitura: lê e exibe o conteúdo
                        conteudo = arquivo.read()
                        print(f"Conteúdo lido:\n{conteudo}")
                    elif modo_escolhido == 'w':
                        # Modo escrita: solicita e escreve novo conteúdo
                        novo_conteudo = input("Digite o novo conteúdo para o arquivo (sobrescreverá o existente): ")
                        arquivo.write(novo_conteudo + "\n")
                        print("Conteúdo escrito no arquivo.")
                    elif modo_escolhido == 'a':
                        # Modo adicionar: solicita conteúdo e adiciona ao final
                        conteudo_para_adicionar = input("Digite o conteúdo para adicionar ao arquivo: ")
                        arquivo.write(conteudo_para_adicionar + "\n")
                        print("Conteúdo adicionado ao arquivo.")

            elif modo_escolhido == 'd':
                # Modo download
                if os.path.exists(nome_arquivo):
                    try:
                        files.download(nome_arquivo)
                        print(f"Arquivo '{nome_arquivo}' solicitado para download.")
                    except Exception as e:
                        print(f"Ocorreu um erro durante o download: {e}")
                else:
                     print(f"Erro: Arquivo '{nome_arquivo}' não encontrado para download.")

            elif modo_escolhido == 'x':
                # Modo excluir
                if os.path.exists(nome_arquivo):
                    try:
                        os.remove(nome_arquivo)
                        print(f"Arquivo '{nome_arquivo}' excluído com sucesso.")
                    except Exception as e:
                        print(f"Ocorreu um erro ao tentar excluir o arquivo: {e}")
                else:
                    print(f"Erro: Arquivo '{nome_arquivo}' não encontrado para exclusão.")


        except FileNotFoundError:
            # Este erro geralmente só ocorre no modo 'r' se o arquivo não existir
            print(f"Erro: Arquivo '{nome_arquivo}' não encontrado no modo de leitura ('r').")
        except IOError as e:
            print(f"Erro de E/S ao manipular o arquivo: {e}")
        except Exception as e:
            print(f"Ocorreu um erro inesperado: {e}")
````

'''O trecho de código  "except Exception as e:" faz parte de um bloco try...except em Python.
Ele é usado para capturar e lidar com quaisquer erros ou exceções inesperadas que possam ocorrer durante a execução do código dentro do bloco try.

except Exception as e: especifica que, se qualquer tipo de exceção (erro) ocorrer no bloco try, ela será capturada. A exceção capturada é atribuída à variável e
pode ser usada para acessar informações sobre o erro (como a mensagem de erro).
print(f"Ocorreu um erro durante o download: {e}"): Esta linha dentro do bloco except será executada se uma exceção for capturada.
Ela imprime uma mensagem informativa indicando que ocorreu um erro durante o download e inclui a mensagem de erro específica contida na variável e.
Em resumo, este bloco garante que, se algo der errado durante a tentativa de download do arquivo (além do arquivo não ser encontrado, que é tratado pelo except FileNotFoundError),
o programa não trave, mas sim capture o erro, imprima uma mensagem e continue a execução (ou encerre de forma controlada).'''

````python
#Exemplo com nível básico de segurança:
#Este código "simulará" a necessidade da senha para ler e escrever no arquivo.

#Passos para usar este exemplo:
#Abra o Gerenciador de Secrets no painel esquerdo (ícone de chave).
#Clique em "Adicionar Secret".
#No campo "Name", digite FILE_PASSWORD.
#No campo "Value", digite a senha que você deseja usar.
#Execute a célula de código que vou gerar abaixo.


from google.colab import userdata
import os

# Define o nome do arquivo
nome_arquivo_secreto = "arquivo_com_secret.txt"

# Obtém a senha do Gerenciador de Secrets do Colab
# Certifique-se de ter adicionado um Secret com o nome 'FILE_PASSWORD'
try:
    senha_secreta = userdata.get('FILE_PASSWORD')
    if senha_secreta is None:
        print("Erro: Secret 'FILE_PASSWORD' não encontrado. Adicione-o no Gerenciador de Secrets.")
except Exception as e:
    print(f"Erro ao acessar o Secret: {e}")
    senha_secreta = None # Define como None se houver erro ao obter o secret


# Função para simular login e manipular o arquivo
def manipular_arquivo_com_senha(senha_fornecida):
    if senha_secreta is None:
        print("\nOperação cancelada: Senha secreta não disponível.")
        return

    # Verifica se a senha fornecida corresponde à senha armazenada no Secret
    if senha_fornecida == senha_secreta:
        print("\nSenha correta. Acessando o arquivo...")
        try:
            # Verifica se o arquivo existe
            if os.path.exists(nome_arquivo_secreto):
                print(f"Arquivo '{nome_arquivo_secreto}' encontrado.")
                # Lê o conteúdo atual
                with open(nome_arquivo_secreto, "r", encoding="utf-8") as arquivo:
                    conteudo = arquivo.read()
                    print(f"Conteúdo atual:\n{conteudo}")

                # Adiciona uma nova linha (simulando escrita)
                with open(nome_arquivo_secreto, "a", encoding="utf-8") as arquivo:
                    nova_linha = f"Esta linha foi adicionada com a senha correta em {os.path.getmtime(nome_arquivo_secreto)}.\n"
                    arquivo.write(nova_linha)
                    print("Nova linha adicionada.")

                # Lê o conteúdo atualizado
                with open(nome_arquivo_secreto, "r", encoding="utf-8") as arquivo:
                    conteudo_atualizado = arquivo.read()
                    print(f"Conteúdo atualizado:\n{conteudo_atualizado}")

            else:
                # Cria o arquivo se não existir
                print(f"Arquivo '{nome_arquivo_secreto}' não encontrado. Criando...")
                with open(nome_arquivo_secreto, "w", encoding="utf-8") as arquivo:
                    arquivo.write("Conteúdo inicial protegido por senha.\n")
                    print(f"Arquivo '{nome_arquivo_secreto}' criado com sucesso.")

        except IOError as e:
            print(f"Erro de E/S ao manipular o arquivo: {e}")
        except Exception as e:
             print(f"Ocorreu um erro inesperado: {e}")

    else:
        print("\nSenha incorreta. Acesso negado.")

````
